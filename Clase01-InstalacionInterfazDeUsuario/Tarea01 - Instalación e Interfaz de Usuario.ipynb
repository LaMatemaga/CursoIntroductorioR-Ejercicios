# Seccin 3: Sobre aspectos visuales de RStudio

Es necesario que hagas alguna de las dos opciones (t decidirs cul) para entregar. El objetivo es saber ajustar los aspectos visuales de la interfaz.

### Opcin A: Cambia el tema de colores de RStudio

**Paso 1.** Haz clic en el men `Tools` de la barra superior y haz clic en la opcin `Global Options...`.

**Paso 2.** Haz clic en el men `Appearance`. Ve a la seccin de `Editor Theme` y navega entre las diferentes opciones. Selecciona la que sea de tu agrado y haz clic en `Apply`.

**Paso 3.** Cierra la ventana, escribe este cdigo y toma una captura de la nueva vista de tu RStudio. No es necesario correrlo, es slo para ver cmo cambi la visualizacin.

```r
# plotting of R objects
plot <- function (x, y, ...)
{
  if (is.function(x) && 
      is.null(attr(x, "class")))
  {
    if (missing(y))
      y <- NULL
    
    # check for ylab argument
    hasylab <- function(...) 
      !all(is.na(
        pmatch(names(list(...)),
              "ylab")))
    
    if (hasylab(...))
      plot.function(x, y, ...)
    
    else 
      plot.function(
        x, y, 
        ylab = paste(
          deparse(substitute(x)),
          "(x)"), 
        ...)
  }
  else 
    UseMethod("plot")
}
```
